```python
# define Global Variables the control the game workflow
tplSelection = ("Rock","Paper","Scissor")
```

```python
# define the function that test user choice against computer choice
import random
def ChoiceTest (UserChoice):
    strComputerChoice = random.choice(tplSelection)
    print(f"My choice is : {strComputerChoice} ")
    strChoiceTest = ""
    if(UserChoice == tplSelection[0] and strComputerChoice == tplSelection[1]):
        strChoiceTest = "Computer"
    elif(UserChoice == tplSelection[0] and strComputerChoice == tplSelection[2]):
        strChoiceTest = "User"
    elif(UserChoice == tplSelection[1] and strComputerChoice == tplSelection[0]):
        strChoiceTest = "User"
    elif(UserChoice == tplSelection[1] and strComputerChoice == tplSelection[2]):
        strChoiceTest = "Computer"
    elif(UserChoice == tplSelection[2] and strComputerChoice == tplSelection[0]):
        strChoiceTest= "Computer"
    elif(UserChoice == tplSelection[2] and strComputerChoice == tplSelection[1]):
        strChoiceTest= "User"
    else:
        strChoiceTest = "Equal"
    return strChoiceTest

#ChoiceTest("Paper")
```

```python
# Game main code
intGameCounter=0
intUserScore = 0
intComputerScore = 0
intLoopCondition=0
print("Welcome on Rock Paper Scissor Game")
print("==================================\n")
print("1- Start Game")
print("2- Exit Game")
intLoopCondition = int(input("Select an option"))
while intLoopCondition == 1:
    intGameCounter +=1
    print("")
    print(f"Game Number: {intGameCounter}")
    print( "______________________________")
    print("1- Rock")
    print("2- Paper")
    print("3- Scissor")
    intUserChoice = int(input("what is your choice? "))
    try:
        strResult = ChoiceTest(tplSelection[intUserChoice-1])
    except:
        print("ERORR!: Please enter numbers from 1 to 3")
        continue
    if(strResult == "User"):
        intUserScore += 1
        print("Concratulations You won")
    elif(strResult == "Computer"):
        intComputerScore += 1
        print("Sorry I won")
    else:
        print("Drawn")
    print("")
    print(f"Your score is: {intUserScore}")
    print(f"My Score is: {intComputerScore}")
    print("")
    print("1- Continue Game")
    print("2- Exit Game")
    intLoopCondition = int(input("Select an option"))
            
```
